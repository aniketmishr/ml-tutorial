```python
import pandas as pd
import numpy as np
import seaborn as sn
import math
from sklearn.model_selection import train_test_split
#dataset link: https://www.kaggle.com/merishnasuwal/breast-cancer-prediction-dataset
```

```python
df = pd.read_csv("Breast_cancer_data.csv")
df.head()
```

```python
X = df[["mean_radius", "mean_texture", "mean_perimeter"]].to_numpy()
y= df["diagnosis"].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

#Gaussian Naive Bayes From sklearn

```python
# training the model on training set
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# making predictions on the testing set
y_pred = gnb.predict(X_test)
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

```python
class CustomGaussianNB:

  def fit(self,X,y):
    self._X = X
    self._y = y

  def predict(self,X):
    pred = []
    for x in X:
      classes = np.unique(self._y)
      posteriors = []
      for y in classes:
        liklihood = self.calculate_liklihood(x,y)
        posteriors.append(liklihood*self.calculate_prior(y))
      pred.append(classes[posteriors.index(max(posteriors))])
    return pred


  def calculate_prior(self,value):
    count = 0
    for item in self._y:
      if (item == value):
        count+=1
    prior = count/len(self._y)
    return prior

  def calculate_liklihood(self,feature_vector,label):
    liklihood_list = []
    rows,columns = self._X.shape
    for i in range(columns):
      column = self._X[:,i]
      labelCol = []
      for k in range(len(column)):
        if self._y[k] == label:
          labelCol.append(column[k])
      mean = np.mean(labelCol)
      std = np.std(labelCol)

      _prob = (1 / (np.sqrt(2 * np.pi) * std)) *  np.exp(-((feature_vector[i]-mean)**2 / (2 * std**2 )))
      liklihood_list.append(_prob)

    liklihood = 1
    for i in liklihood_list:
      liklihood *=i

    return liklihood
```

#Custom Gaussian Naive Bayes Implementation from scratch

```python
customModel = CustomGaussianNB()
customModel.fit(X_train,y_train)
customModelOutput = customModel.predict(X_test)
print("Custom Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, customModelOutput)*100)
```
